```python
import cv2
import time
import matplotlib.pyplot as plt
```

```python
print(cv2)
```

```python
image = cv2.imread("test_img.jpg" , 1)
print(image.shape)
```

```python
resized = cv2.resize(image , (900,700))
cv2.imshow("Palm" , resized)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
# by default its bgr convert it to rgb
```

```python
video = cv2.VideoCapture(0)     # capture first frame
check , frame = video.read()
time.sleep(3)
cv2.imshow("Capturing" , frame)
cv2.waitKey(0)
video.release()
cv2.destroyAllWindows()
```

```python
# capture live video using for loop
video = cv2.VideoCapture(0)     # capture first frame
frames = 1;

while True:
    frames = frames + 1
    check, frame = video.read()
    cv2.imshow("Live Capture" , frame)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break;

print(frames)
video.release()
cv2.destroyAllWindows()
    
```
